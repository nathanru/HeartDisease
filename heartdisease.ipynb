###Which one should you use (pyplot vs matplotlib 00 method?)
* When plotting something quickly, ok to use pyplot method
* When plotting something more advance, use the OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
#ax.set_xlim([45,100]);
```

```python
fix, ax = plt.subplots(figsize=(10,6))

#plot the data

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

#customize the plot 

ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

#Add a legend

ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle='--',color='r');
```

```python
over_50.head()
```

```python
#subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=False)
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(),
          linestyle='--',color='r');

#--------------------------------------------------------
# plot for ax1
scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);

ax1.set(title="Heart Disease and Max Heart Rate(Thalach) ",
        xlabel="Age",
        ylabel="Max Heart Rate (Thalach)");

ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(),
          linestyle='-',color='g');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");




#-------------------------------------------------------
# hist = ax2.hist(over_50["chol"])

# ax2.set(title="Histogram of Cholesterol Levels over age 50");

# ax2.axhline(over_50["chol"].mean(),
#           linestyle='-',color='g');

```

```python
plt.style.use("seaborn-whitegrid")

fix, ax = plt.subplots(figsize=(10,6))

#plot the data

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); #this changes the color scheme

#customize the plot 

ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

#Add a legend

ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle='--',color='r');
```

```python
#customizing the x and y limitations

#subplot of chol, age, thalach
#plot for ax0
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=False)
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");
ax0.set_xlim([50,80])

ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(),
          linestyle='--',color='r');

#--------------------------------------------------------
# plot for ax1
scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                     cmap="winter");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.set(title="Heart Disease and Max Heart Rate(Thalach) ",
        xlabel="Age",
        ylabel="Max Heart Rate (Thalach)");

ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(),
          linestyle='--',color='r');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");
```
